```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/RAW_DATA_ALL (3).csv')
```

```python

df
```

```python
#statistcal summary
```

```python
plt.figure(figsize=(5,5))
plt.subplot(1,2,1)
sns.histplot(df.Rating)
plt.subplot(1,2,2)
sns.histplot(df.ratings)
```

```python
df1=df[df.Rating>4.5]
df1.sort_values(by='Rating',ascending=False,inplace=True)
df1
```

```python
#Ratings>4.5,Top 10 rated book
plt.xticks(rotation=90)
plt.yticks(ticks=np.arange(0,5,.5))
sns.barplot(x=df1.book_title,y=df1.Rating,data=df1,orient='v').set_title('Top 10 rated books')
```

```python
#No of rating > 400000
df2=df[df.ratings>40000]
df2.sort_values(by='ratings',ascending=False,inplace=True)
df2.head(20)
df2.ratings.sort_values(ascending=False)
df2x=df2.head(10)
df2x
```

```python
plt.xticks(rotation=90)
sns.barplot(x='book_title',y='ratings',data=df2x,orient='v').set_title('Top highly reviewed 10 books')
```

```python
df3=df.author_id.value_counts()
df3_re=df3.reset_index()
df3_re.columns=['author_id','count']
df3_re.author_id.astype('str')
df3_re1=df3_re.sort_values(by='count',ascending=False)
df3_re1
```

```python
df4=df3_re1.head(5)
df4.author_id=df4.author_id.astype('str')
```

```python

sns.barplot(x='author_id',y='count',data=df4,color='red').set_title('Top 5 Publishing Author id')
```
