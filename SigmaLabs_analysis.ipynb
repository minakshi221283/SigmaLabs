```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/RAW_DATA_ALL (3).csv')
```

```python

df
```

```python
#statistcal summary
```

```python
plt.figure(figsize=(5,5))
plt.subplot(1,2,1)
sns.histplot(df.Rating)
plt.subplot(1,2,2)
sns.histplot(df.ratings)
```

```python
df1=df[df.Rating>4.5]
df1
```

```python
#Ratings>3.5,Top 10 rated book

sns.barplot(y=df1.book_title,x=df1.Rating,data=df1,orient='h').set_title('Top 10 rated books')
```

```python
#No of rating > 40000
df2=df[df.ratings>40000]
```

```python
sns.barplot(y=df1.book_title,x=df1.ratings,data=df2,orient='h').set_title('Top highly reviewed 10 books')
```

```python
df3=df.author_id.value_counts()
df3_re=df3.reset_index()
df3_re.columns=['author_id','count']
df3_re.author_id.astype('str')
df3_re1=df3_re.sort_values(by='count',ascending=False)
df3_re1
```

```python
df4=df3_re1.head(5)
df4.author_id=df4.author_id.astype('str')
```

```python

sns.barplot(x='author_id',y='count',data=df4,color='red').set_title('Top 5 Authors')
```
